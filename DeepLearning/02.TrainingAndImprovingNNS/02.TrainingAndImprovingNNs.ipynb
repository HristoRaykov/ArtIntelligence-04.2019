```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy, SparseCategoricalCrossentropy
from tensorflow.keras.metrics import Accuracy, SparseCategoricalAccuracy

from tensorflow.keras.datasets import mnist
```

```
tf.__version__
```

# Training And Improving Neural Networks

```
(train_attributes, train_labels), (test_attributes, test_labels) = mnist.load_data()
train_attributes, test_attributes = train_attributes / 255.0, test_attributes / 255.0
```

```
model = Sequential([
                    Input(shape=(28, 28)), 
                    Flatten(),
                    Dense(20, activation="relu", kernel_regularizer="l2"), # bias_regularizer="l2", activity_regularizer="l2"
                    Dropout(0.2),
                    Dense(10, activation="relu", kernel_regularizer="l2"),
                    Dropout(0.2),
                    Dense(10, activation="softmax", kernel_regularizer="l2")
])
model.summary()
```

```
model.layers[1].bias.numpy()
```

```
model.layers[1].kernel.numpy().shape
```

```
model.layers[1].weights
```

```
model.compile(optimizer=Adam(learning_rate=0.001), loss=SparseCategoricalCrossentropy(), metrics=[SparseCategoricalAccuracy()])# "accuracy"
```

```
model.fit(train_attributes, train_labels, epochs=10, batch_size=16)
```

```
print(model.predict_classes(train_attributes[:10]))
print(train_labels[:10])
```
